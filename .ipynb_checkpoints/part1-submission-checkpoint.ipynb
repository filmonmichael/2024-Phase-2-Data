```python
print("Unique Categories: ", len(Sales_data['Category'].unique()))
print("Unique Sub- Categories: ", len(Sales_data['Sub-Category'].unique()))
print("Unique Regions: ", len(Sales_data['Region'].unique()))
print("Unique Ship Mode: ",len(Sales_data['Ship Mode'].unique()))
print("Unique States: ", len(Sales_data['State'].unique()))
print("Unique Cities: ",len(Sales_data['City'].unique())) 
print("Unique Countries: ",len(Sales_data['Country'].unique()))
print("Unique Segments: ",len(Sales_data['Segment'].unique()))
```

Category and Country have 1 unique value therefore they're dropped because they don't add information. City contains a large number of unique values which means that it's not suitable for one-hot encoding. Hence the variables to be one-hot encoded are Sub-Category, Regions, Ship Mode, State, Segment.

```python
columns_to_keep = ['Sub-Category', 'Region', 'State', 'Ship Mode', 'Sales', 'Quantity', 'Discount', 'Profit', 'Segment', 'Order Date']
Reduced_Sales_data = Sales_data[columns_to_keep]

# Apply one-hot encoding to relevant columns
categorical_columns = ['Sub-Category', 'Region', 'State', 'Ship Mode', 'Segment']
Sales_data_encoded = pd.get_dummies(Reduced_Sales_data, columns=categorical_columns, drop_first=False) # can also drop_first = True however since the number of new vectors is relatively low we can keep them for visualisation
Sales_data_encoded.head()
```

```python
# Summing or aggregating numeric features by Order Date
numeric_aggregation = Sales_data_encoded.groupby('Order Date').agg({
    'Sales': 'sum',
    'Quantity': 'sum',
    'Discount': 'mean',
    'Profit': 'sum'
}).reset_index()

# Summing one-hot encoded variables by Order Date
one_hot_aggregation = Sales_data_encoded.groupby('Order Date').sum().reset_index()
one_hot_aggregation.drop(columns=['Sales', 'Quantity','Discount','Profit'], inplace=True)
# Ensuring the date columns are aligned correctly
aggregated_df = pd.merge(numeric_aggregation, one_hot_aggregation, on='Order Date', how='left')
aggregated_df
```

Looking at the first 5 rows we can see that some dates are skipped. Perhaps no orders were made on those dates. I have aggregated the data further to be weekly and monthly data.

```python
aggregated_df.set_index('Order Date', inplace=True) # must set the order date as the index for it to be aggregated correctly
```

```python
# Downsampling to weekly, monthly frequency, summing the sales for each month

# Resampling the data to monthly frequency, however Discount is a ratio so it should be the mean instead of sum

weekly_resampled_df = aggregated_df.resample('W').agg({
    'Discount': 'mean',
    
    # One-hot encoded columns should be summed
    **{col: 'sum' for col in aggregated_df.columns if col not in ['Discount']}
})


monthly_resampled_df = aggregated_df.resample('M').agg({
    
    'Discount': 'mean',
    
    # One-hot encoded columns should be summed
    **{col: 'sum' for col in aggregated_df.columns if col not in [ 'Discount']}
})

weekly_resampled_df.head()
```

```python
weekly_resampled_df.shape
```

```python
weekly_resampled_df.describe()
```

After aggregating the data, checking the shape of the data shows that we now have weekly data for years 2014 to 2017 with no skipped week in between which makes it suitable for timeseries analysis. Weekly data is being to use for this anlaysis because it provides more observations 208 compared to monthly 48. This is so that the model has sufficient data to learn the underlying patterns and make accurate predictions. 

```python
# correlation matrix
correlation_matrix = weekly_resampled_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

It seems that alot of the one hot encoded variables are highly correlated to each other. Some represent categories within the same overarching categorical feature (like different states within regions and different sub-categories within a category). High correlations (close to 1 or -1) indicate redundancy or overlap in the information they provide. This can lead to unstable estimates in regression models. Quantity seems to be highly correlated with sales and must be dropped. Variables with high correlations between each other will have one of the two dropped.

```python
# Identifying outliers using IQR method
Q1 = weekly_resampled_df['Sales'].quantile(0.25)
Q3 = weekly_resampled_df['Sales'].quantile(0.75)
IQR = Q3 - Q1
outliers = weekly_resampled_df[(weekly_resampled_df['Sales'] < (Q1 - 1.5 * IQR)) | (weekly_resampled_df['Sales'] > (Q3 + 1.5 * IQR))]

# Optionally, cap the outliers
weekly_resampled_df['Sales'] = np.where(weekly_resampled_df['Sales'] > (Q3 + 1.5 * IQR), Q3 + 1.5 * IQR, weekly_resampled_df['Sales'])
weekly_resampled_df['Sales'] = np.where(weekly_resampled_df['Sales'] < (Q1 - 1.5 * IQR), Q1 - 1.5 * IQR, weekly_resampled_df['Sales'])

outliers
```

We can see that with IQR method there were no outliers. However we will check the residuals after fitting model to see any unusual values.

## Feature Selection

```python
#Feature selection to help reduce the correlation between the variables
corr_matrix = weekly_resampled_df.corr().abs()
# Selecting upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Finding features with correlation greater than a threshold of 0.8
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]

# Dropping features
weekly_reduced_df = weekly_resampled_df.drop(columns=to_drop)
```

```python
#Feature selection to help reduce the correlation between the variables
corr_matrix = monthly_resampled_df.corr().abs()
# Selecting upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Finding features with correlation greater than a threshold 0.9
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

# Dropping features
monthly_red_cols = monthly_resampled_df.drop(columns=to_drop)
```


## 5. Summary

To preprocess the data, I initially checked the data type of the variables and whether there were missing values in our dataset. This data contained categorical and continuous data with no missing values. The categorical variables were one hot encoded so that they could be interpreted by our model. I have narrowed the categorical variables to Sub-Category, Regions, Ship Mode, State, and Segment. These variables were the most suitable to be one-hot encoded and provided valuable information for our model to learn the patterns in our data. 

To identify patterns such as trend and seasonality, I aggregated the data into weekly and monthly frequencies. Weekly data provides more data points for our model to learn from hence this was the preferred grain for analysis. Discount is a ratio therefore the mean was used instead of the sum to aggregate it. Although visually there appeared to be outliers these were not removed as they could be captured later in the model's seasonal terms. Furthermore using the IQR method showed that we did not have significant outliers. 
After identifying highly correlated features, I dropped one out of the pairs to reduce multi-colinearity and create stable estimates through our model. 
I plotted the one-hot encoded data in bar plots against the weekly, monthly, and yearly periods.
After the one-hot encoding of the variables, I attempted to use Principal Component (PCA) analysis to reduce the dimensionality while maintaining the majority of the variance in the data. I was able to reduce the dimensions to 22 Principal Components. However, this did not work well for our data because PCA does not take into account the fact that time series data is time ordered (time indexed). PCA maximizes variance but it would make the interpretability of our model difficult. Hence, I used the correlations instead. 

Some variables had wider or smaller ranges than others. For example, the Discount variable had a smaller range when compared to other variables such as Sales in the dataset. If the model was fit on the data as it was, it could have been biased towards variables with larger numbers. To reduce this and make it easier for the model to identify the patterns in the data and I standardized and scaled the data. I saved my preprocessed data into a CSV file to make it easy to access.
